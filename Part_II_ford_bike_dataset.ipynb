## Investigation Overview


The primary goal of this presentation is to explore and analyze GoBike data to uncover insights and trends that can help us understand usage patterns, user demographics, and other key factors influencing bike share services. By delving into univariate, bivariate, and multivariate analyses, we aim to provide a comprehensive overview of the dataset and identify actionable insights.



## Dataset Overview and Executive Summary

The dataset contains information about bike share rides, including duration, start and end times, start and end station details, user demographics, and bike share status. It consists of approximately 183,412 entries with 16 features.

After doing explorative analysis of our dataset, we found the following insights:
1. The distribution of ride durations follows a skewed pattern, with most rides lasting between a few minutes to around one hour. However, there are outliers with significantly longer durations.
2. Rides are more frequent during peak hours, particularly around 8 AM in the morning and 5 PM in evening, suggesting commuter usage patterns.
3. Subscribers constitute the majority of users, indicating a strong base of regular users, while customers represent occasional users.
4. Male users outnumber female and other gender users, indicating potential gender-based disparities in bike share usage.
5. Most rides do not involve bike sharing for the entire trip.
6. There is a wide range of birth years among users, with the majority falling between the 1980 to 2000.

These insights provide a foundation for further exploration and analysis, enabling us to make informed decisions regarding bike share service management, marketing strategies, and infrastructure development.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
df=pd.read_csv('201902-fordgobike-tripdata.csv')
````

## Visualization 1: Distribution of Ride Durations by Member Birth Year

````python
# Box plot illustrating the distribution of ride durations for different member birth years. 
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='member_birth_year', y='duration_sec', color='skyblue')
plt.title('Distribution of Ride Durations by Member Birth Year')
plt.xlabel('Member Birth Year')
plt.ylabel('Duration (seconds)')
plt.xticks(rotation=90)
plt.show()
````

This plot helps identify any potential trends or outliers in ride durations across different age groups. It showed that there is no clear linear relationship between ride duration and member birth year. Most ride durations are concentrated within a certain range across all member birth years.

## Visualization 2: Distribution of Ride Durations by Start Hour

````python
df['start_hour'] = pd.to_datetime(df['start_time']).dt.hour
# Box plot displaying the distribution of ride durations for different start hours of the day.
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='start_hour', y='duration_sec', color='salmon')
plt.title('Distribution of Ride Durations by Start Hour')
plt.xlabel('Start Hour')
plt.ylabel('Duration (seconds)')
plt.show()
````

This plot reveals variations in ride durations based on the time of day when the ride starts. It revealed that ride durations vary across different start hours. Longer ride durations tend to occur during certain hours of the day, while shorter ride durations are more common during other hours.

## Visualization 3: Distribution of Ride Durations by User Type

````python
# Box plot demonstrating how ride durations differ between different user types (Customers and Subscribers).
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='user_type', y='duration_sec', palette='Set2')
plt.title('Distribution of Ride Durations by User Type')
plt.xlabel('User Type')
plt.ylabel('Duration (seconds)')
plt.show()
  
````

This plot helps understand whether there are significant differences in ride durations based on user type and showed that there are differences in ride durations between different user types. Subscribers generally have shorter ride durations compared to customers, as indicated by the lower median and quartiles in the box plot.

## Visualization 4: Distribution of Ride Durations by Member Gender

````python
# Box plot showcasing the distribution of ride durations across different member genders.
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='member_gender', y='duration_sec', palette='pastel')
plt.title('Distribution of Ride Durations by Member Gender')
plt.xlabel('Member Gender')
plt.ylabel('Duration (seconds)')
plt.show()
````

This plot provides insights into potential differences in ride durations based on gender. This revealed that there are some differences in ride durations between different member genders.However, the differences are not as pronounced as those observed between user types.

## Visualization 5: Count of Rides by User Type and Member Gender

````python
# clustered bar chart illustrating the count of rides categorized by user type and member gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='user_type', hue='member_gender', palette='viridis')
plt.title('Count of Rides by User Type and Member Gender')
plt.xlabel('User Type')
plt.ylabel('Count')
plt.legend(title='Member Gender')
plt.show()
````

This plot helps compare the distribution of rides among different user types and genders. It showed that distribution of user types varies across different member genders.For both males and females, subscribers are more common compared to customers.The proportion of subscribers is higher among males compared to females.

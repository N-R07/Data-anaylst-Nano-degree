```python
import sqlite3
import pandas as pd
```

```python
#Connect to the SQLite database. Here, database was created as it didn't exists.
conn = sqlite3.connect('movie_database.db')
```

```python
# Loading csv
df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Writing the DataFrame 'df' to a SQLite database table named 'movie_credits'. It will create a table in this case as no table before.
df.to_sql('movie_credits', conn, if_exists='replace', index=False)
```

```python
# connection close
conn.close()
```

```python
conn = sqlite3.connect('movie_database.db')
# Example SQL query to select all data from the table
query = "SELECT * FROM movie_credits"

# Executing query and reading into DataFrame
df = pd.read_sql_query(query, conn)

# Displaying the DataFrame
df
```

The columns in both the dataframes which are stored in the JSON string format will be cleaned in the tidiness issue.

### **Tidiness Issue 1:  Multiple Variables Stored in One Column**

```python
# Parsing the 'genres' column to a list of dictionaries
df1_clean['genres'] = df1_clean['genres'].apply(lambda x: json.loads(x.replace('\'', '"')) if isinstance(x, str) else [])

# Exploding the 'genres' list to create separate rows for each genre
df1_clean_exploded = df1_clean.explode('genres')

# Extracting genre names (as each dictionary in 'genres' list has a 'name' key)
df1_clean_exploded['genre'] = df1_clean_exploded['genres'].apply(lambda x: x.get('name') if isinstance(x, dict) else None)

# Dropping the original 'genres' column if no longer needed
df1_clean_exploded = df1_clean_exploded.drop(columns=['genres'])
```

```python
#FILL IN - Validate the cleaning was successful
df1_clean_exploded.head()
```

```python
# Parsing the stringified JSON data in the 'cast' column
df2_credit['cast'] = df2_credit['cast'].apply(lambda x: json.loads(x) if isinstance(x, str) else [])

# Exploding the 'cast' list to create separate rows for each cast member
df2_clean_exploded_cast = df2_credit.explode('cast')

# Extracting the cast name from the list of dictionaries in the 'cast' column
df2_clean_exploded_cast['cast_name'] = df2_clean_exploded_cast['cast'].apply(lambda x: x.get('character') if isinstance(x, dict) else None)

# Dropping the original 'cast' column if no longer needed
df2_clean_exploded = df2_clean_exploded_cast.drop(columns=['cast'])
```

```python
#Validating that cast column has also been exploded

df2_clean_exploded.head()
```

Justification: According to tidy data principles, each variable must have its own column, and each observation must have its own row. By parsing and exploding the genres and cast column, we adhere to these principles. This will also allow us to examine how individual genres or cast member perform affect success metrics without the complexities of dealing with multi-genre movies or multiple cast members as a single observation. I have not parsed crew, keyword and other similar columns as they are not required for analysis and so I'll remove them before merging datasets.

### **Tidiness Issue 2: Inconsistent Column Names Across DataFrames:**

```python
# Validate the cleaning by checking column names
df2_clean_exploded.columns.tolist()
```

The movie_id column in df2_clean_exploded has been successfully renamed to id, aligning with the column name in df1_clean_exploded.

Depending on the datasets, you can also peform the combination before the cleaning steps.

### **5.1:** Define and answer the research question 
Going back to the problem statement in step 1, use the cleaned data to answer the question you raised. Produce **at least** two visualizations using the cleaned data and explain how they help you answer the question.

*Research question:* What is the correlation between various movie attributes, such as budget, genres, and cast/crew information, with indicators of movie success like revenue and ratings.

Defining function to avoid repetitive code

```python
# Function to create scatter plots to explore correlations
def plot_correlation(df, x_var, y_var, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=x_var, y=y_var)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

####  Visual 1 -Correlation between Budget and Revenue

```python
# Correlation between Budget and Revenue
plot_correlation(combined_df, 'budget', 'revenue', 'Correlation between Budget and Revenue', 'Budget', 'Revenue')
```

*Answer to research question:* The scatter plot showing the correlation between budget and revenue suggests that there is a positive relationship between these two variables. Movies with higher budgets tend to generate higher revenue, which can be attributed to larger budgets allowing for better production quality, more prominent actors, and more extensive marketing campaigns that attract larger audiences.

#### Visual 2 - Correlation between Budget and Ratings

```python
# Correlation between Budget and Ratings
plot_correlation(combined_df, 'budget', 'vote_average', 'Correlation between Budget and Ratings', 'Budget', 'Ratings')
```

*Answer to research question:* The correlation between the budget of a movie and its ratings (vote_average) does not show a strong linear relationship as seen with revenue. While there are movies with high budgets and high ratings, there's a wide spread of ratings across different budget levels. This suggests that a larger budget does not necessarily guarantee higher ratings.
